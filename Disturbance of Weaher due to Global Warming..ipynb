7. Data Cleaning


```python
df1['Temperature (C)'] = df1['Temperature (C)'].astype('float')
df1['Apparent Temperature (C)'] = df1['Apparent Temperature (C)'].astype('float')
df1['Humidity'] = df1['Humidity'].astype('float')
df1 [ 'Wind Speed (km/h)'] = df1[ 'Wind Speed (km/h)'].astype('float')
df1 [ 'Wind Bearing (degrees)'] = df1[ 'Wind Bearing (degrees)'].astype('int')
df1 [ 'Visibility (km)'] = df1['Visibility (km)'].astype('float')
df1['Loud Cover'] = df1['Loud Cover'].astype('int')
df1['Pressure (millibars)'] = df1['Pressure (millibars)'].astype('float')
```

```python
df1['Precip Type'].fillna( df1['Precip Type'].value_counts().index [0], inplace= True )
df1.isnull().sum()
```
